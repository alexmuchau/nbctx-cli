# Teste 0

```python
import math
from statistics import mean
```

## Teste 1

Blalbal

```python
numbers = [3, 5, 8, 13, 21]
scale = 2
```

```python
scaled = [value * scale for value in numbers]
roots = [round(math.sqrt(value), 2) for value in scaled]
```

```python
offset = 10
adjusted = [value + offset for value in scaled]
```

### Teste 3

```python
summary = {
    "count": len(numbers),
    "average_scaled": mean(scaled),
    "average_adjusted": mean(adjusted),
    "largest_root": max(roots),
}
```

```python
print(f"Processed {summary['count']} values")
print(summary)
```

```python
print("Adjusted values:", adjusted)
print("Summary with adjusted average:", summary)
```
